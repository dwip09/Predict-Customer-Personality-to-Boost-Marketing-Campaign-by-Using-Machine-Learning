## Conversion rate analysis based on income, spending and age

##### Menghapus nilai 0 pada kolom `NumWebVisitsMonth`

```python
df[(df.NumWebVisitsMonth == 0)].shape[0]
```

```python
df = df[~(df.NumWebVisitsMonth == 0)]
```

### Conversion Rate Analysis Based on Age

```python
df['Dt_Customer'] = df['Dt_Customer'].apply(lambda x : x.split('-')[-1])

df['Age'] = np.int64(df['Dt_Customer']) - df['Year_Birth']
df['Age_Category'] = df['Age'].apply(lambda x : 'Dewasa' if x < 40 else ('Lansia' if x < 60 else ('Manula')))
```

```python
dfg_age = df.groupby(['Age_Category']).agg({'Response':'count'}).reset_index()
dfg1 = df[df.Response == 1].groupby(['Age_Category']).agg({'Response':'count'}).reset_index()
dfg1['Conversion Rate'] = (dfg1['Response'] / dfg_age['Response']) * 100

dfg1['Age_Category'] = pd.Categorical(dfg1['Age_Category'], categories=['Dewasa', 'Lansia', 'Manula'])
dfg1 = dfg1.sort_values('Age_Category')
dfg1
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x=dfg1['Age_Category'], y='Conversion Rate', data=dfg1)

plt.title('Conversion Rate Analysis Based on Age', weight='bold', fontsize = 18)
plt.tight_layout()

plt.bar_label(ax.containers[0], fmt='%.2f%%')
plt.show()
```

### Conversion Rate Analysis Based on Income

```python
df['Income_Category'] = df['Income'].apply(lambda x : 'Low' if x < 35000000 else ('Mid' if x < 55000000 else ('High')))
```

```python
dfg_income = df.groupby(['Income_Category']).agg({'Response':'count'}).reset_index()
dfg2 = df[df.Response == 1].groupby(['Income_Category']).agg({'Response':'count'}).reset_index()
dfg2['Conversion Rate'] = (dfg2['Response'] / dfg_income['Response']) * 100

dfg2['Income_Category'] = pd.Categorical(dfg2['Income_Category'], categories=['Low', 'Mid', 'High'])
dfg2 = dfg2.sort_values('Income_Category')
dfg2
```

```python
plt.figure(figsize=(13,6))
ax = sns.barplot(x=dfg2['Income_Category'], y='Conversion Rate', data=dfg2)

plt.title('Conversion Rate Analysis Based on Income', weight='bold', fontsize = 18)
plt.tight_layout()

plt.bar_label(ax.containers[0], fmt='%.2f%%')
plt.show()
```

### Conversion Rate Analysis Based on Spending

```python
df['Total_Spending'] = df['MntCoke'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['Spending_Category'] = df['Total_Spending'].apply(lambda x : 'Low' if x < 150000 else ('Mid' if x < 1000000 else ('High')))
```

```python
dfg_spending = df.groupby(['Spending_Category']).agg({'Response':'count'}).reset_index()
dfg3 = df[df.Response == 1].groupby(['Spending_Category']).agg({'Response':'count'}).reset_index()
dfg3['Conversion Rate'] = (dfg3['Response'] / dfg_spending['Response']) * 100

dfg3['Spending_Category'] = pd.Categorical(dfg3['Spending_Category'], categories=['Low', 'Mid', 'High'])
dfg3 = dfg3.sort_values('Spending_Category')
dfg3
```

```python
plt.figure(figsize=(13,6))
ax = sns.barplot(x=dfg3['Spending_Category'], y='Conversion Rate', data=dfg3)

plt.title('Conversion Rate Analysis Based on Spending', weight='bold', fontsize = 18)
plt.tight_layout()

plt.bar_label(ax.containers[0], fmt='%.2f%%')
plt.show()
```

### Feature Transformation

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# StandardScaler
df_fs['Recency_scale'] = StandardScaler().fit_transform(df_fs['Recency'].values.reshape(len(df_fs),1))

# MinMaxScaler
df_fs['Total_Purchase_scale'] = MinMaxScaler().fit_transform(df_fs['Total_Purchase'].values.reshape(len(df_fs),1))
df_fs['Total_Spending_scale'] = MinMaxScaler().fit_transform(df_fs['Total_Spending'].values.reshape(len(df_fs),1))
```

```python
df_scale = df_fs.drop(['Recency', 'Total_Purchase', 'Total_Spending'], axis=1)
```

# Modeling

## Clustering (K-Means)

### Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scale)
    inertia.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(2, 11), y=inertia, color='blue', linewidth = 3)
sns.scatterplot(x=range(2, 11), y=inertia, s=100, color='red')
plt.show()
```

```python
(pd.Series(inertia) - pd.Series(inertia).shift(-1)) / pd.Series(inertia) * 100
```

### Silhouette Score

```python
from sklearn.metrics import silhouette_score

range_n_clusters = list(range(2,11))
sil_score_euclidean = []
for c in range_n_clusters:
    kmeans = KMeans(n_clusters=c).fit(df_scale)
    prediction = kmeans.predict(df_scale)

    score_euclidean = silhouette_score(df_scale, prediction, metric='euclidean')
    sil_score_euclidean.append(score_euclidean)
```

```python
sns.lineplot(x=range(2, 11), y=sil_score_euclidean, color='blue', linewidth = 3)
sns.scatterplot(x=range(2, 11), y=sil_score_euclidean, s=100, color='red')
plt.show()
```

### Hasil Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42).fit(df_scale)
clusters = kmeans.labels_
df_encode['clusters'] = clusters
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df_scale)
Xpca = pca.transform(df_scale)
```

```python
df_pca = pd.DataFrame(data = Xpca, columns = ['pc1', 'pc2'])
```
